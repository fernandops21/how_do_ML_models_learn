```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

plt.style.use("seaborn-v0_8-darkgrid")
```

```python
x = 2 * np.random.rand(100, 1) # numeros entre 0 e 1
y = 4 + 3 * x + np.random.randn(100, 1)
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.scatter(x, y);
```

Y_hat = theta_0_hat + theta_1_hat * x

Queremos encontrar theta_0_hat e theta_1_hat que minimizam a função custo (MSE). Ou seja, ao calcularmos a distancia de cada ponto pra reta, obtermos as menores distâncias possíveis (em um mundo ideal, e hipotético, todos os pontos pontos estiriam sobre a reta e essa soma de distancias seria 0).

Solução analítica:
Theta_hat = (X_t * X)^(-1) * X_t * y

```python
X_b = np.c_[np.ones((100,1)), x]

theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta
```

```python
y_hat = theta[0] + theta[1] * x

fig, ax = plt.subplots(figsize=(20, 7))
plt.plot(x, y_hat)
plt.scatter(x, y);
```

### Gradient Descent

Algorimo iterativo  

Theta_next = theta_atual - eta * gradient_MSE_theta  
Onde: gradient_MSE_theta é o vetor de derivadas parciais do MSE em função dos thetas  

eta = learning rate (tamanho "do passo")  


gradient_MSE_theta = 2/m * X_t * (X * theta_atual - y)

```python
eta = 0.1
n_iterations = 10
m = 100

theta = np.random.randn(2,1)
fig, ax = plt.subplots(figsize=(25, 8))
plt.scatter(x, y)

for iteration in range(n_iterations):
    grandient = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * grandient

    y_hat = theta[0] + theta[1] * x

    ax.plot(x, y_hat, alpha=0.1 + iteration/n_iterations, color='red')
```
